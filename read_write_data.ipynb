```python
import csv
import pandas as pd
import re
from preprocessing import clean, save_numeric_data
import numpy as np
csv.field_size_limit(10000000)
```

## Loading data

```python
articles = []
with open('articles.csv', 'r', encoding="utf8") as file:
    reader = csv.reader(file, delimiter=',')
    for r in reader:
        try:
            article = r[1]
            articles.append(article)
        except:
            pass
        try:
            article = r[2]
            articles.append(article)
        except:
            pass
print(len(articles), 'articles ready')
```

## Dropping too small articles

```python
articles = pd.Series(articles)
lens = pd.Series([len(a) for a in articles])
articles_new = articles[lens > 1000]
print(articles_new.shape)
```

## Saving

```python
articles_new.to_csv('articles.csv')
```

## Dropping first paragraph (first paragraphs are too similar)

```python
articles = []
with open('articles.csv', 'r', encoding="utf8") as file:
    reader = csv.reader(file, delimiter=',')
    for r in reader:
        article = r[1]
        articles.append(article)
    articles = articles[1:]
print(len(articles), 'articles ready')
```

```python
new_articles = []
for a in articles:
    art = ''.join(a.split('\n')[1:])
    new_articles.append(art)
```

```python
pd.Series(new_articles).to_csv('articles.csv')
```

## Dropping too small ones ( < 700 characters)

```python
data = []
with open('articles.csv', 'r', encoding="utf8") as file:
    reader = csv.reader(file, delimiter=',')
    for r in reader:
        article = r[1]
        data.append(article)
    data = data[1:]
    data = pd.Series(data)
print(data.shape, 'articles ready')
```

```python
data = data[data.apply(len) > 700].apply(clean)
print(data.shape)
```

## Making a list of all paragraphs

```python
paragraphs = []
for t in data:
    paragraphs.extend(t.split('<PARAGRAPHEND>'))
paragraphs = [p+'<PARAGRAPHEND>' for p in paragraphs if (len(p) > 300)]
print(len(paragraphs))
```

## Making a list of bi-grams of sentences from each paragraph

```python
sentences2 = []
for t in paragraphs:
    txt = [sent+'<SENTENCEEND>' for sent in t.split('<SENTENCEEND>') if len(sent)>50]
    if len(txt) >= 2:
        res = []
        sents = []
        for l in range(len(txt)-1):
            sents.append(''.join(txt[l:l+2]))
        res.extend(sents)
        sentences2.extend(res)
sentences2 = [s for s in sentences2 if (len(s) < 300)]
print(len(sentences2))
```

## Saving a file

```python
with open('data_final/2sentences.txt', 'w', encoding="utf8") as file:
    for value in sentences2:
        file.write(value+'\n')
```

## Adding similar data from media articles to sentences2.txt

```python
with open('2sentences_media.txt', 'r', encoding="utf8") as file:
    for value in file:
        sentences2.append(value[:-1])
len(sentences2)
```

## Dropping mid-sentence <PARAGRAPHEND>-s

```python
for txt in sentences2:
    txt.replace('<PARAGRAPHEND>', '')
```

## Saving

```python
with open('data_final/2sentences.txt', 'w', encoding="utf8") as file:
    for value in sentences2:
        file.write(value+'\n')
```
